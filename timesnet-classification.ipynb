```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/working/Time-Series-Library/test.csv').T[1:]
class_names = ['F0L', 'F0M', 'F1L', 'F1M', 'F2L', 'F2M', 'F3L', 'F3M', 'F4L', 'F4M', 'F5L', 'F5M', 'F6L', 'F6M', 'F7L', 'F7M']

y_pred = df[0]
y_true = df[1]
```

```python
np.mean(y_pred == y_true)
```

```python
print(classification_report(y_true,y_pred, target_names = class_names))
```

```python
c = confusion_matrix(y_true,y_pred)
df_cm = pd.DataFrame(c, index = class_names, columns = class_names)
plt.figure(figsize = (20,20))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
sn.heatmap(df_cm, annot=True)
```
